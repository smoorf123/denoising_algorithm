### Gaussian, Mean and Mode Denoising

```python
'''These forms of denoising are outdated and perform relatively poorly compared to more modern methods.
Only used here for testing purposes.'''

import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv

from skimage import io
from scipy import ndimage as nd

img = io.imread('test2.jpeg')
g_img = nd.gaussian_filter(img, sigma=1) # gaussian image using a sigma value of 2
m_img=nd.median_filter(img, size=4) # median filtering using sampling window of size 4

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,15))

ax1.imshow(img), ax1.axis('off') # noisy image
ax2.imshow(g_img), ax2.axis('off') # denoised image - Gaussian Filtering
ax3.imshow(m_img), ax3.axis('off') # denoised image - Median Filtering

plt.show()

''' Gaussian Filtering does a decent job at removing noise however edge detection and clarity is low.
Median Filtering is not much better in terms of overall clarity of denoised image hoever edge detection
and clarity is better than the Gaussian filter '''
```

### Non-Local Means Denoising

```python
'''Takes a small "window" (say 5x5) around a noisy pixel and averages multiple other
similar looking "windows" in neighbouring areas and replaces original noisy pixel using a weighted average.'''

import cv2 as cv

img = cv.imread('test2.jpeg') # noisy image

dst = cv.fastNlMeansDenoisingColored(img,None,10,10,7,21) # input src, output dest, filter strength, filter strength for color imgs, size of template window, size of search window

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,10))

# converting BGR channel images from cv2 to RGB for matplotlib; turning off axis for all images
ax1.imshow(img), ax1.axis('off') # noisy image
ax2.imshow(dst), ax2.axis('off') # denoised image

plt.show()
```
